```python
# Load tokenizer and model from Hugging Face
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
model = AutoModel.from_pretrained('bert-base-uncased')

# Our sentence
sentence = "The Transformer has revolutionized NLP."

# Tokenize the sentence
inputs = tokenizer(sentence, return_tensors='pt')
input_ids = inputs['input_ids']

# Print out the tokens and their corresponding IDs
print('Token IDs:', input_ids)

# Get the embeddings for the input IDs
with torch.no_grad(): # We don't need gradients for this operation
    outputs = model(input_ids)
    embeddings = outputs.last_hidden_state

# Choose a token to display its embedding, for example, the first token
token_id = input_ids[0][1]  # Get the token ID for the second token (index 1)
token_embedding = embeddings[0][1]  # Get the embedding for the second token

print(f"Embedding for token ID {token_id}:", token_embedding)
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

# Load tokenizer and model from Hugging Face
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
model = AutoModel.from_pretrained('bert-base-uncased')

# Our sentence
sentence = "The Transformer has revolutionized NLP."

# Tokenize the sentence
inputs = tokenizer(sentence, return_tensors='pt')

# Get the input IDs and pass them through the model to get the embeddings
input_ids = inputs['input_ids']
outputs = model(input_ids)

# The last hidden state is the embedding of the tokens
embeddings = outputs.last_hidden_state
print(embeddings)

# Accessing the positional embeddings (indirectly)
# Note: This is more of an illustrative purpose. For specific models like BERT, positional embeddings are not directly exposed.
# However, you can access the model's embeddings layer and look at the weights for the positional embeddings.
positional_embeddings = model.embeddings.position_embeddings.weight
print(positional_embeddings)

# If you want to see the effect of positional embeddings on a specific token,
# you can compare the embeddings with and without positional information.
# Note: This is a conceptual demonstration and may not directly apply as is.
token_without_position = model.embeddings.word_embeddings(input_ids[0])
token_with_position = embeddings[0]

print("Token without positional information:", token_without_position)
print("Token with positional information:", token_with_position)
```
